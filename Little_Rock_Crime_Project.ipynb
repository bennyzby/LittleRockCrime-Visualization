```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://xiaoxl.github.io/Datasets/contents/assests/datasets/lrpd.csv')
```

```python
# Remove all the duplicates in the column INCIDENT_NUMBER.
df.drop_duplicates(subset='INCIDENT_NUMBER', inplace=True)
```

```python
# Set INCIDENT_NUMBER to be the index to uniquely identify offenses.
df.set_index('INCIDENT_NUMBER', inplace=True)
```

```python
df
```

```python
# Fill the missing values on WEAPON_TYPE to be NO WEAPON. Fill all other missing values by 0.
df['WEAPON_TYPE'].fillna('NO WEAPON', inplace=True)
df.fillna(0, inplace=True)
```

```python
df
```

```python
# Change 1 in WEAPON_TYPE to be UNKNOWN.
df['WEAPON_TYPE'].replace(1, 'UNKNOWN', inplace=True)
df
```

```python
# Reassigned the columns to their respective types, e.g. numbers are numeric types and characters are strings.
df = df.astype({'LOCATION_DISTRICT':str, 'OFFENSE_CODE':str,'OFFENSE_DESCRIPTION':str,'WEAPON_TYPE':str,'INCIDENT_LOCATION':str,'CITY':str,'STATE':str,'ZIP':int,'LATITUDE':float,'LONGITUDE':float,'Location 1':str,'Offense Status':str})
```

```python
# Parse the INCIDENT_DATE to find out the year, month, day, hour and min and create columns accordingly. Note that AM/PM is taken into consideration when computing hours.
df['INCIDENT_DATE'] = pd.to_datetime(df['INCIDENT_DATE'], format='%m/%d/%Y %I:%M:%S %p')
df['YEAR'] = df['INCIDENT_DATE'].dt.year
df['MONTH'] = df['INCIDENT_DATE'].dt.month
df['DAY'] = df['INCIDENT_DATE'].dt.day
df['HOUR'] = df['INCIDENT_DATE'].dt.hour
df['MINUTE'] = df['INCIDENT_DATE'].dt.minute
df['DayOfWeek'] = df['INCIDENT_DATE'].dt.dayofweek
```

```python
# Create a column CRIME_TYPE based on the following rules:
def crime_type(offense_code, offense_description):
    if '23' in str(offense_code):
        return 'Non-Violent'
    elif 'THEFT' in offense_description:
        return 'Non-Violent'
    else:
        return 'Violent'

df['CRIME_TYPE'] = df.apply(lambda x: crime_type(x['OFFENSE_CODE'], x['OFFENSE_DESCRIPTION']), axis=1)
```

```python
# Create a column RISK_TYPE based on the following rules:
def risk_type(crime_type, weapon_type):
    if crime_type == 'Non-Violent' and weapon_type == 'NO WEAPON':
        return 'Low Risk'
    else:
        return 'High Risk'

df['RISK_TYPE'] = df.apply(lambda x: risk_type(x['CRIME_TYPE'], x['WEAPON_TYPE']), axis=1)
```

```python
# Counts of crimes on each day of a week, per crime type
sns.set_theme()
sns.set(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'gray'})
df_weekday = df.groupby(['CRIME_TYPE', 'DayOfWeek'])['YEAR'].count().reset_index()
plt.figure(figsize=(20,8))
sns.barplot(data=df_weekday, x='DayOfWeek', y='YEAR', hue='CRIME_TYPE')
plt.xlabel('Day of Week')
plt.ylabel('Counts of Crimes')
plt.show()
```

```python
# Counts of crimes on each month of a year, per crime type
df_month = df.groupby(['CRIME_TYPE', 'MONTH'])['YEAR'].count().reset_index()
plt.figure(figsize=(20,8))
sns.barplot(x='MONTH', y='YEAR', hue='CRIME_TYPE', data=df_month)
plt.title('Counts of Crimes by Month and Crime Type')
plt.xlabel('Month')
plt.ylabel('Counts of Crimes')
plt.show()
```

```python
# Counts of crimes on each part of a day, per crime type
bins = [-1, 6, 12, 18, 24]
labels = ['0-6', '6-12', '12-18', '18-24']
df['TIME_PERIOD'] = pd.cut(df['HOUR'], bins=bins, labels=labels, include_lowest=True)
df_time = df.groupby(['CRIME_TYPE', 'TIME_PERIOD'])['YEAR'].count().reset_index()
plt.figure(figsize=(20,8))
sns.barplot(x='TIME_PERIOD', y='YEAR', hue='CRIME_TYPE', data=df_time)
plt.title('Counts of Crimes by Time of Day and Crime Type')
plt.xlabel('Time of Day')
plt.ylabel('Counts of Crimes')
plt.show()
```

```python
# Counts of crimes on each part of a day, per risk type
df_risk = df.groupby(['RISK_TYPE', 'TIME_PERIOD'])['YEAR'].count().reset_index()
plt.figure(figsize=(20,8))
sns.barplot(x='TIME_PERIOD', y='YEAR', hue='RISK_TYPE', data=df_risk)
plt.title('Counts of Crimes by Time of Day and Risk Type')
plt.xlabel('Time of Day')
plt.ylabel('Counts of Crimes')
plt.show()
```
